```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import pandas as pd

# Configuration
BATCH_SIZE = 64
EPOCHS = 20
LEARNING_RATE = 0.001
INPUT_SHAPE = (32, 32, 1)  # Redimensionné de 28x28 à 32x32 pour ResNet
NUM_CLASSES = 10

# Labels Fashion-MNIST
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

def preprocess_data():
    """Charge et préprocesse les données Fashion-MNIST"""
    # Charger Fashion-MNIST
    (x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()
    
    # Normaliser les pixels entre 0 et 1
    x_train = x_train.astype('float32') / 255.0
    x_test = x_test.astype('float32') / 255.0
    
    # Redimensionner de 28x28 à 32x32 et ajouter dimension de canal
    x_train = tf.image.resize(x_train[..., np.newaxis], [32, 32])
    x_test = tf.image.resize(x_test[..., np.newaxis], [32, 32])
    
    # Convertir labels en categorical
    y_train = keras.utils.to_categorical(y_train, NUM_CLASSES)
    y_test = keras.utils.to_categorical(y_test, NUM_CLASSES)
    
    return (x_train, y_train), (x_test, y_test)

def residual_block(x, filters, stride=1, conv_shortcut=False):
    """Bloc résiduel de base pour ResNet"""
    shortcut = x
    
    # Première couche convolutionnelle
    x = layers.Conv2D(filters, 3, strides=stride, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    
    # Deuxième couche convolutionnelle
    x = layers.Conv2D(filters, 3, padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    
    # Shortcut connection
    if conv_shortcut:
        shortcut = layers.Conv2D(filters, 1, strides=stride, use_bias=False)(shortcut)
        shortcut = layers.BatchNormalization()(shortcut)
    
    # Addition et activation finale
    x = layers.Add()([shortcut, x])
    x = layers.Activation('relu')(x)
    
    return x

def build_resnet18(input_shape, num_classes):
    """Construit le modèle ResNet-18"""
    inputs = keras.Input(shape=input_shape)
    
    # Couche initiale
    x = layers.Conv2D(64, 7, strides=2, padding='same', use_bias=False)(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D(3, strides=2, padding='same')(x)
    
    # Blocs résiduels - Couche 1 (64 filtres)
    x = residual_block(x, 64)
    x = residual_block(x, 64)
    
    # Blocs résiduels - Couche 2 (128 filtres)
    x = residual_block(x, 128, stride=2, conv_shortcut=True)
    x = residual_block(x, 128)
    
    # Blocs résiduels - Couche 3 (256 filtres)
    x = residual_block(x, 256, stride=2, conv_shortcut=True)
    x = residual_block(x, 256)
    
    # Blocs résiduels - Couche 4 (512 filtres)
    x = residual_block(x, 512, stride=2, conv_shortcut=True)
    x = residual_block(x, 512)
    
    # Classification head
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dropout(0.5)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    
    model = keras.Model(inputs, outputs, name='resnet18')
    return model

def plot_training_history(history):
    """Visualise les courbes d'apprentissage"""
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # Loss
    axes[0, 0].plot(history.history['loss'], label='Train Loss', color='blue')
    axes[0, 0].plot(history.history['val_loss'], label='Validation Loss', color='red')
    axes[0, 0].set_title('Training and Validation Loss')
    axes[0, 0].set_xlabel('Epochs')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].legend()
    axes[0, 0].grid(True)
    
    # Accuracy
    axes[0, 1].plot(history.history['accuracy'], label='Train Accuracy', color='blue')
    axes[0, 1].plot(history.history['val_accuracy'], label='Validation Accuracy', color='red')
    axes[0, 1].set_title('Training and Validation Accuracy')
    axes[0, 1].set_xlabel('Epochs')
    axes[0, 1].set_ylabel('Accuracy')
    axes[0, 1].legend()
    axes[0, 1].grid(True)
    
    # Learning rate (si utilisé un scheduler)
    if 'lr' in history.history:
        axes[1, 0].plot(history.history['lr'], label='Learning Rate', color='green')
        axes[1, 0].set_title('Learning Rate Schedule')
        axes[1, 0].set_xlabel('Epochs')
        axes[1, 0].set_ylabel('Learning Rate')
        axes[1, 0].set_yscale('log')
        axes[1, 0].legend()
        axes[1, 0].grid(True)
    else:
        axes[1, 0].axis('off')
    
    # Résumé des métriques finales
    final_train_acc = history.history['accuracy'][-1]
    final_val_acc = history.history['val_accuracy'][-1]
    final_train_loss = history.history['loss'][-1]
    final_val_loss = history.history['val_loss'][-1]
    
    axes[1, 1].text(0.1, 0.8, f'Final Train Accuracy: {final_train_acc:.4f}', fontsize=12)
    axes[1, 1].text(0.1, 0.7, f'Final Val Accuracy: {final_val_acc:.4f}', fontsize=12)
    axes[1, 1].text(0.1, 0.6, f'Final Train Loss: {final_train_loss:.4f}', fontsize=12)
    axes[1, 1].text(0.1, 0.5, f'Final Val Loss: {final_val_loss:.4f}', fontsize=12)
    axes[1, 1].text(0.1, 0.3, f'Overfitting: {final_train_acc - final_val_acc:.4f}', fontsize=12)
    axes[1, 1].set_title('Training Summary')
    axes[1, 1].axis('off')
    
    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred, class_names):
    """Visualise la matrice de confusion"""
    cm = confusion_matrix(y_true, y_pred)
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Matrice de confusion absolue
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=class_names, yticklabels=class_names, ax=axes[0])
    axes[0].set_title('Confusion Matrix (Absolute)')
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('Actual')
    
    # Matrice de confusion normalisée
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names, ax=axes[1])
    axes[1].set_title('Confusion Matrix (Normalized)')
    axes[1].set_xlabel('Predicted')
    axes[1].set_ylabel('Actual')
    
    plt.tight_layout()
    plt.show()
    
    return cm

def plot_sample_predictions(model, x_test, y_test, class_names, num_samples=12):
    """Visualise quelques prédictions du modèle"""
    predictions = model.predict(x_test[:num_samples])
    predicted_classes = np.argmax(predictions, axis=1)
    true_classes = np.argmax(y_test[:num_samples], axis=1)
    
    fig, axes = plt.subplots(3, 4, figsize=(12, 9))
    axes = axes.ravel()
    
    for i in range(num_samples):
        img = x_test[i].squeeze()
        axes[i].imshow(img, cmap='gray')
        
        predicted = class_names[predicted_classes[i]]
        actual = class_names[true_classes[i]]
        confidence = predictions[i][predicted_classes[i]]
        
        color = 'green' if predicted_classes[i] == true_classes[i] else 'red'
        axes[i].set_title(f'Pred: {predicted}\nActual: {actual}\nConf: {confidence:.2f}', 
                         color=color, fontsize=10)
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.show()

def main():
    print("🚀 Début de l'entraînement ResNet-18 sur Fashion-MNIST")
    print("=" * 60)
    
    # 1. Préparation des données
    print("📊 Chargement et préprocessing des données...")
    (x_train, y_train), (x_test, y_test) = preprocess_data()
    
    print(f"Train shape: {x_train.shape}")
    print(f"Test shape: {x_test.shape}")
    print(f"Number of classes: {NUM_CLASSES}")
    
    # 2. Construction du modèle
    print("\n🏗️ Construction du modèle ResNet-18...")
    model = build_resnet18(INPUT_SHAPE, NUM_CLASSES)
    
    # Compilation du modèle
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=LEARNING_RATE),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
    print(f"Nombre de paramètres: {model.count_params():,}")
    
    # 3. Callbacks pour l'entraînement
    callbacks = [
        keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
        keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=3, min_lr=1e-7),
        keras.callbacks.ModelCheckpoint('best_resnet18.h5', save_best_only=True)
    ]
    
    # 4. Entraînement
    print(f"\n🎯 Début de l'entraînement ({EPOCHS} epochs)...")
    history = model.fit(
        x_train, y_train,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        validation_data=(x_test, y_test),
        callbacks=callbacks,
        verbose=1
    )
    
    # 5. Évaluation finale
    print("\n📈 Évaluation finale...")
    train_loss, train_acc = model.evaluate(x_train, y_train, verbose=0)
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)
    
    print(f"Train Accuracy: {train_acc:.4f}")
    print(f"Test Accuracy: {test_acc:.4f}")
    print(f"Train Loss: {train_loss:.4f}")
    print(f"Test Loss: {test_loss:.4f}")
    
    # 6. Visualisations
    print("\n📊 Génération des visualisations...")
    
    # Courbes d'apprentissage
    plot_training_history(history)
    
    # Prédictions pour matrice de confusion
    y_pred = model.predict(x_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true_classes = np.argmax(y_test, axis=1)
    
    # Matrice de confusion
    cm = plot_confusion_matrix(y_true_classes, y_pred_classes, class_names)
    
    # Rapport de classification
    print("\n📋 Rapport de classification:")
    print(classification_report(y_true_classes, y_pred_classes, 
                              target_names=class_names))
    
    # Exemples de prédictions
    plot_sample_predictions(model, x_test, y_test, class_names)
    
    # 7. Analyse par classe
    print("\n🎯 Performance par classe:")
    class_accuracy = cm.diagonal() / cm.sum(axis=1)
    for i, (class_name, acc) in enumerate(zip(class_names, class_accuracy)):
        print(f"{class_name:15}: {acc:.4f}")
    
    print("\n✅ Entraînement terminé!")
    
    return model, history

# Exécution
if __name__ == "__main__":
    model, history = main()
```
